```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

col=['Id' , 'Entity' , 'Sentiment' , 'Content']
data=pd.read_csv(r"D:\My Data\Downloads\archive\twitter_training.csv",names=col)
data.head()
```

```python
sentiment_count=data.Sentiment.value_counts()
sentiment_count
```

```python
y=['Neutral' , 'Negative' , 'Positive','Irrelevant']
plt.pie(sentiment_count , labels=y, autopct='%0.1f%%' )
circle=plt.Circle((0,0),0.4, facecolor='white')
plt.gca().add_patch(circle)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
entity_counts = data['Entity'].value_counts()
entity_counts
```

```python
top_10_entity = entity_counts.head(10)
top_10_entity
```

```python
# Bar Chart representation of top 10 most engaged entities on tweets

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

# Plot a bar graph for the top 10 categories with colors
plt.figure(figsize=(10, 6))
top_10_entity.plot(kind='bar', color=colors)

# Remove the annotate part to not show values on top of each bar

plt.xlabel("Entities")
plt.ylabel('Tweet Count')
plt.title('Top 10 Entities with Highest Tweet Engagement')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.show()
```

```python
#Setting up the plot
plt.figure(figsize=(15, 7))

#Plotting entity distribution for training data
sns.countplot(data, y=data.columns[1], order=data[data.columns[1]].value_counts().index, palette='viridis')
plt.title('Entity Distribution in Training Data')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.tight_layout()
plt.show()
```
